Code for printing image:

```python
def imshow(img):
    import cv2
    import IPython
    _,ret = cv2.imencode('.jpg', img) 
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
```

```python
imshow(img)
```

```python
small_img = cv2.resize(img, (224,224))
```
